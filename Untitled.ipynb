```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
import pickle
pd.options.display.max_rows = 100
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
slots = {"food":1, "wood":2,"clay":3,"stone":4,"gold":5,"card1":6,"card2":7,"card3":8,"card4":9,"prod":10,"tent":11}
df = df.applymap(lambda x: slots[x] if x in slots else x)
df
```

```python
df.columns
```

```python
df.corr()
```

```python
cors = df.corr().abs()
cors.values[np.triu_indices_from(cors, 0)] = np.nan
cors = cors.unstack()
cors = cors.sort_values(na_position="first")
```

```python
df.corr()["fitness"].abs().sort_values()
```

```python
X = df.drop("fitness",axis=1)
y = df["fitness"]
encoder = preprocessing.LabelEncoder()
y_enc = encoder.fit_transform(y)
```

```python
model = KNeighborsClassifier(n_neighbors = 100).fit(X,y_enc)
```

```python
f = open("KNN100model","wb")
pickle.dump(model,f)
f.close()
```
